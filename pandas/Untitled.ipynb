```python
# import xlwings as xw
import pandas as pd
path = {'bigdata' : "C:\\Users\\Dmitrii\\Desktop\\pandas\\БД ТС Omnicomm.xlsx",
       'ab10' : "C:\\Users\\Dmitrii\\Desktop\\pandas\\Сводный табель АБ №10-8 МАРТ 2021.xlsx",
       'um10' : "C:\\Users\\Dmitrii\\Desktop\\pandas\\Сводный табель УМ №10-8 МАРТ 2021.xlsx",
       'uchastock41' : "C:\\Users\\Dmitrii\\Desktop\\pandas\\Март 2021 дт.xlsx",
       'ab44' : '',
       'um44' : '',
       'po' : "C:\\Users\\Dmitrii\\Desktop\\pandas\\omnicomm.xlsx"}
data_bd = pd.read_excel(path['bigdata']) # открываем БД
data_bd = data_bd[data_bd.klad.notnull()] # отфильтровываем избыточные значения с None

data_ab_10 = pd.read_excel(path['ab10']) # откываем АБ
data_ab_10.drop(data_ab_10.columns.difference(['Unnamed: 2', 'Unnamed: 3']), 1, inplace=True) # берем 2 колонки: название ТС и заправку
data_ab_10.columns = ['klad', 'g'] # переименоввываем столбцы

data_um_10 = pd.read_excel(path['um10']) # открываем УМ
data_um_10.drop(data_um_10.columns.difference(['Unnamed: 4', 'Unnamed: 5']), 1, inplace=True) # берем 2 колонки: название ТС и заправку
data_um_10.columns = ['klad', 'g'] # переименоввываем столбцы

data_41 = pd.read_excel(path['uchastock41'])
data_41.drop(data_41.columns.difference(['Unnamed: 1', 'Unnamed: 3']), 1, inplace=True) # берем 2 колонки: название ТС и заправку
data_41.columns = ['klad', 'g'] # переименоввываем столбцы

df = pd.concat([data_ab_10, data_um_10, data_41])
df1 = pd.merge(data_bd, df)
df1 = df1[df1.g.notnull()]
```

```python
# редактируем сводный отчет
```

```python
df1 = df1.dropna(axis='index', how='any', subset=['g', 'klad']).drop('klad', axis=1)
data3 = pd.read_excel("C:\\Users\\Dmitrii\\Desktop\\pandas\\omnicomm.xlsx")
data3 = data3.drop([0, 1, 2, 3, 4, 5, 6]) #удалить первых 7 строк
data3 = data3.drop(['Unnamed: 6', 'Unnamed: 7'], axis = 1) # удаляемем ненужные колонки
```

```python
result = pd.merge(data3, df1, left_on = 'Unnamed: 0', right_on = 'Omnicomm').drop('Omnicomm', axis=1).drop('Unnamed: 4', axis=1)
```
